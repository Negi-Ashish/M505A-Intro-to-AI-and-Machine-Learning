```python
import nbformat

notebook_path = "final project.ipynb"  # change if needed

with open(notebook_path, "r", encoding="utf-8") as f:
    nb = nbformat.read(f, as_version=4)

total_chars = 0
markdown_chars = 0
code_chars = 0

for cell in nb.cells:
    cell_len = len(cell.source)

    total_chars += cell_len

    if cell.cell_type == "markdown":
        markdown_chars += cell_len
    elif cell.cell_type == "code":
        code_chars += cell_len

print(f"Total Characters: {total_chars:,}")
print(f"Markdown Characters: {markdown_chars:,}")
print(f"Code Characters: {code_chars:,}")
```
